# Copy

```python
d1 = {"one": 1, "Two": 2, 3: 3, 1+2j: 4}
d1
```

# Fromkeys

```python
d3 = dict.fromkeys(d1)
d3
```

```python
val = 10
d4 = dict.fromkeys(d1,val)
d4
```

```python
keys = { 'a', 'b', 'c', 3, 8j}
d5 = dict.fromkeys(keys,val)
d5
```

```python
values = [10,20,30]
d6= dict.fromkeys(keys, values)
d6
```

```python
values.append(40)
d6
```
