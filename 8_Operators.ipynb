## Bitwise Operator

- complement (~)
- AND (&)
- OR (|)
- XOR (^)
- Left Shift <<
- Right Shift >>


```python
~12  # It does binary negation
```

```python
~(-13)
```

```python
12 & 13 #1100 & 1101
```

```python
12 | 13
```

```python
12 ^ 13  # True if 1 and 0 or 0 and 1; otherwise false
```

```python
100 << 3  #simple calculation: multiply given number with 2 to the power of given number (100 * 2 ** 3)
```

```python
800 >> 3  #simple calculation: divide given number with 2 to the power of given number (100 / 2 ** 3)
```
